Distance in megaparsecs (Mpc) on the x-axis, and Velocity in kilometers per second (km/s) on the y-axis

```python
import numpy as np
import matplotlib.pyplot as plt #always start with numpy
```

```python
X = np.array([19,65,66,80,97,113,175,270,310,350,350,650,680])    #importing data x being the distnace in mpc
Y = np.array([1150,3800,5000,4800,5400,6700,10300,15400,19500,23200,21600,39400,41000])  #and y being the speed in km/s
```

```python
plt.plot(X,Y,'g.') #plotting our data
```

```python
m_guess = 0
b_guess = 0 #this tells the system that we do not know what m or b is yet
```

$y=mx+b$  (basic slope formula)

```python
m_guess * X + b_guess #creates a new system of numbers
```

```python
y_guess = m_guess * X + b_guess #trying to see how well our model fits our data
```

```python
y_guess - Y
```

```python
np.sum((y_guess - Y)**2) #squaring it so our negative numbers into postives aka getting rid of the negatives
```

```python
loss = np.sum((y_guess - Y)**2)
```

```python
best_loss_so_far = 1000000000000 #finds the smallest loss out of all the possibilites
m_best = 0 
b_best = 0 
for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):
        
        y_guess = m_guess * X + b_guess #the system will now try to guess y using a number of posibilities from m and b by comparing it to the model

        loss = np.sum((y_guess - Y)**2)

        if loss < best_loss_so_far: #if the loss is better than anything we have seen so far the system now replace it
            m_best = m_guess
            b_best = b_guess
            best_loss_so_far = loss

print("Found best fit:")
print(m_best,b_best,loss) #now we know the slope, y intecept, and loss  we can now guess what is the best fit
```

```python
m_best
```

```python
H = m_best #h is the hubble constant whihc is the rate of expansion of the universe
```

```python
1/H
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name):
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df #importing hubble law calculations into colab
```

```python
url = 'https://docs.google.com/spreadsheets/d/1B_HW2uVGQv9W7b6vVZDSkOPLtmfodHrX4G0qtQbIenk/edit?usp=sharing' #distance to a galaxy calculations
```
