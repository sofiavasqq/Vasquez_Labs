# Pie Chart

```python
labels = 'Frogs', 'Dogs', 'Cats', 'Birds' #controls what each part of the chart is labeled
sizes = [15, 30, 45, 10] #controls how large the percent of each label is in the graph
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Dogs') #controls how far apart you want to the chart to be(0 being the closest)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90) #controls if you want a shadow or not
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show() #displays the pie chart
```

### Functions

```python
def square(x):
    return x**2 #this function will square whatever number is input square(?)
```

```python
square(5) #as seen here, square(5) equals 25
```

```python
def add_two(x):
    return x+2 #this function will add 2 to whatever is input add_two(?)
```

```python
add_two(5) #as seen here, add_two(5) does 5+2 which equals 7
```

```python
square(add_two(5)) #by using both of the functions created you can create complex equations like the one below which basically is (5+2)^2
```

```python
def coin():
  return np.random.randint(2) #gives us any whole number >2 #tells the system to make a integer up to 2, but not 2 itself
```

```python
display(HTML('<img src="https://random-ize.com/coin-flip/us-quarter/us-quarter-front.jpg">'))
```

```python
heads = 'https://random-ize.com/coin-flip/us-quarter/us-quarter-front.jpg'
tails = 'https://random-ize.com/coin-flip/us-quarter/us-quarter-back.jpg' #inputing our images into the system
```

```python
def show_img(img):
  display(HTML('<img src=' + img + '>')) #check if the image works
```

```python
show_img(heads) #displays the head image that was imported earlier
```

```python
show_img(tails) #displays the tails image that was imported earlier
```

```python
def flip_coin():

    if coin():
        show_img(heads)
    else:
        show_img(tails) #create a function that will allow the coin to show both sides depending on the circumstances
```

```python
flip_coin() #flip coin is finished!
```

```python
for i in range(5):
    flip_coin() #flips the coin five times, giving a different result everytime
```
